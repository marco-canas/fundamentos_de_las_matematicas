## ¬øQu√© es?  

Flipped Learning o aprendizaje invertido es un enfoque pedagoÃÅgico en el que:  

* la instruccioÃÅn directa se realiza fuera del aula y

Aqu√≠ tienes un ejemplo de una ecuaci√≥n financiera com√∫nmente utilizada para calcular el valor futuro de una inversi√≥n con inter√©s compuesto:

La f√≥rmula del valor futuro (FV) de una inversi√≥n con inter√©s compuesto es:

\[ FV = PV \times (1 + r)^n \]

Donde:
- FV es el valor futuro de la inversi√≥n.
- PV es el valor presente o la cantidad inicial de dinero invertida.
- r es la tasa de inter√©s por per√≠odo, expresada en decimal (por ejemplo, 5% se representa como 0.05).
- n es el n√∫mero de per√≠odos de tiempo durante los cuales se invertir√° el dinero.

Por ejemplo, supongamos que invertimos $1,000 en una cuenta de ahorros con una tasa de inter√©s anual del 5% y lo dejamos invertido durante 3 a√±os. Podemos calcular el valor futuro de la inversi√≥n utilizando la ecuaci√≥n financiera:

\[ FV = 1000 \times (1 + 0.05)^3 \]

Calculando el valor futuro:

\[ FV = 1000 \times (1.05)^3 = 1000 \times 1.157625 = 1157.63 \]

El valor futuro de la inversi√≥n despu√©s de 3 a√±os ser√≠a de $1,157.63. Esto significa que con inter√©s compuesto, nuestra inversi√≥n inicial de $1,000 ha crecido a $1,157.63 despu√©s de 3 a√±os con una tasa de inter√©s del 5%.

```python
import pandas as pd 


class Estudiante:
    def __init__(self, nombre, notas):
        self.nombre = nombre
        self.notas = notas

    def promedio(self):
        return sum(self.notas) / len(self.notas)

class GestorEstudiantes:
    def __init__(self):
        self.estudiantes = []

    def agregar_estudiante(self, estudiante):
        self.estudiantes.append(estudiante)

    def obtener_promedio_general(self):
        promedios = [estudiante.promedio() for estudiante in self.estudiantes]
        return sum(promedios) / len(promedios)

    def obtener_promedio_estudiante(self, nombre):
        for estudiante in self.estudiantes:
            if estudiante.nombre == nombre:
                return estudiante.promedio()
        return None

# Crear el gestor de estudiantes
gestor = GestorEstudiantes()

# Agregar estudiantes
estudiante1 = Estudiante("Juan", [4, 4.2, 3.9])
estudiante2 = Estudiante("Mar√≠a", [4, 4.2, 3.9])
estudiante3 = Estudiante("Pedro", [4, 4.2, 3.9])
gestor.agregar_estudiante(estudiante1)
gestor.agregar_estudiante(estudiante2)
gestor.agregar_estudiante(estudiante3)

# Obtener promedio general
promedio_general = gestor.obtener_promedio_general()
print("Promedio general de todos los estudiantes:", promedio_general)

# Obtener promedio de un estudiante espec√≠fico
nombre_estudiante = "Juan"
promedio_estudiante = gestor.obtener_promedio_estudiante(nombre_estudiante)
if promedio_estudiante is not None:
    print("Promedio de", nombre_estudiante + ":", promedio_estudiante)
else:
    print("No se encontr√≥ al estudiante", nombre_estudiante)
```
